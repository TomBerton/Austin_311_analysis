```python
#import Dependencies
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#main dataframe: get the total complaints by zipcode from cleaned 311 data from master
complaints_df = pd.read_csv('raw_data/zipcode_complaints.csv')
del complaints_df["Unnamed: 0"]
complaints_df.head()
```

```python
#Read data for crime by zipcode
crime_df = pd.read_csv('raw_data/crime.csv')
del crime_df["Unnamed: 0"]
crime_df.head()
```

```python
#first merge
zipmain_df = pd.merge(complaints_df, crime_df, on = "zip_code", how = "outer")
zipmain_df.head()
```

```python
zipmain_df = zipmain_df.rename(columns={"complaint count":"complaint_count", "crime reports":"crime_reports"})
zipmain_df.head()
```

```python
#Read data from 2014 household dataset
df_2014 = pd.read_csv('raw_data/2014_household_dataset.csv')
del df_2014["Unnamed: 0"]
df_2014 = df_2014.rename(columns={"Zip Code":"zip_code", "Median household income($)":"median_household_income($)",
                                    "Population below poverty level(%)":"population_below_poverty_level(%)",
                                    "Unemployment(%)":"unemployment(%)"})
df_2014.head()
```

```python
#second merge
zipmain_df = pd.merge(zipmain_df, df_2014, on = "zip_code", how = "outer")
zipmain_df.head()
```

```python
#Read data for population by zipcode
pop_df = pd.read_csv('raw_data/ZIPs_2010_population.csv')
pop_df = pop_df.rename(columns={" male":"male", " female":"female", "total_population_x":"total_population"})
pop_df.head()
```

```python
#Remove commas & convert columns from string to float
pop_df["male"] = pop_df["male"].str.replace(",","").astype(float)
pop_df["female"] = pop_df["female"].str.replace(",","").astype(float)
pop_df["total_population"] = pop_df["total_population"].str.replace(",","").astype(float)

#Calculate male/femae percentage
pop_df["male(%)"] = pop_df.apply(lambda x: (float(x["male"]) / float(x["total_population"])*100), axis=1)
pop_df["female(%)"] = pop_df.apply(lambda x: (float(x["female"]) / float(x["total_population"])*100), axis=1)

#remove male/female column
del pop_df["male"]
del pop_df["female"]
pop_df.head()
```

```python
#third merge
zipmain_df = pd.merge(zipmain_df, pop_df, on = "zip_code", how = "outer")
zipmain_df.to_csv("zip_code1.csv")
```
