```python
#import Dependencies
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#create austin crime df
df_2014 = pd.read_csv('2014_Housing_Market_Analysis_Data_by_Zip_Code.csv',low_memory=False)
df_2014.head()
```

```python
#clean-up data
df_2014 = df_2014.dropna(how="any")
df_2014.shape
```

```python
df_2014["Median household income($)"] = df_2014["Median household income"].map(lambda x: x.lstrip("$"))
del df_2014["Median household income"]
df_2014["Population below poverty level(%)"] = df_2014["Population below poverty level"].map(lambda x: x.rstrip("%"))
del df_2014["Population below poverty level"]
df_2014["Unemployment(%)"] = df_2014["Unemployment"].map(lambda x: x.rstrip("%"))
del df_2014["Unemployment"]
df_2014.head()
```

```python
df_2014_dataset = df_2014[["Zip Code","Median household income($)", "Population below poverty level(%)","Unemployment(%)"]]
df_2014_dataset.reset_index()
df_2014_dataset.head()
```
